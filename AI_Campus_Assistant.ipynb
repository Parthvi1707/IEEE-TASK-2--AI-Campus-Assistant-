# AI Campus Assistant

This project implements a basic AI assistant to help students with campus-related queries
using a pre-trained language model.

```python
from transformers import pipeline

assistant = pipeline(
    "text2text-generation",
    model="google/flan-t5-small"
)
```

```python
def campus_assistant(user_question):
    prompt = (
        "You are a helpful AI campus assistant. "
        "Answer the student's question clearly and politely.\n\n"
        f"Question: {user_question}\nAnswer:"
    )

    response = assistant(prompt, max_length=200)
    return response[0]['generated_text']
```

```python
print(campus_assistant("Where is the college library located?"))
```

```python
print(campus_assistant("What should I do if I miss a class?"))
```

## How It Works

- The user enters a campus-related question.
- The question is sent to a pre-trained language model.
- The model generates a natural language response.
- The response is displayed back to the user.

This forms a basic AI assistant suitable for answering general campus queries.
